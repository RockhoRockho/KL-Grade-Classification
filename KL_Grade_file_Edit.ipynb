```python
import pandas as pd
import os
import shutil
data = pd.read_csv('./data/grade/KL_Grade.csv', header=0, encoding='cp949')
df = pd.DataFrame(data)
df
```

```python
for i in range(len(df)):
    if df['File_name_pic'][i][0:10] == df['real_xml'][i][0:10]:
        pass
    elif df['name'][i][0] != df['temp'][i][0]:
        print('이상있음 : {}' .format(df['name'][i]))
```

```python
df.info()
```

```python
df=df.drop_duplicates(['File_name_pic'])
```

```python
df.info()
```

```python
df.to_csv('./data/grade/KL_grade1st.csv', index=False) # 저장하기
```
